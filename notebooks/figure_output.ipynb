```python
def plot_data(datasets, f1_cg, f1_macro, title_suffix):
      

    barWidth = 0.25
    r1 = np.arange(len(f1_cg))
    r2 = [x + barWidth for x in r1]

    plt.figure(figsize=(15, 10))

    # Calculate reduction percentages
    #f1_cg_reduction = [(1 - (score / f1_cg[0])) * 100 for score in f1_cg]
    #f1_macro_reduction = [(1 - (score / f1_macro[0])) * 100 for score in f1_macro]
    # Calculate reduction percentages
     # Calculate absolute reduction percentages
     # Calculate absolute reduction percentages
        # Calculate the absolute reduction percentages
    f1_cg_reduction = [(1 - (score / f1_cg[0])) * 100 for score in f1_cg]
    f1_macro_reduction = [(1 - (score / f1_macro[0])) * 100 for score in f1_macro]



    # Create bars
    bars1 = plt.bar(r1, f1_cg, width=barWidth, edgecolor='grey', label='F1 (CG) Mean', color='b')
    bars2 = plt.bar(r2, f1_macro, width=barWidth, edgecolor='grey', label='F1 (Macro) Mean', color='r')

    # Title & Subtitle
   
    plt.xlabel('Datasets', fontsize=22)
    plt.ylabel('Scores and Reduction %', fontsize=22)

    # X axis and Y axis tick font size
    plt.xticks([r + barWidth for r in range(len(f1_cg))], datasets, fontsize=16)
    plt.yticks(fontsize=14)
    
        # Before plotting, compute the max reduction value to adjust y-axis
    max_reduction = max(max(f1_cg_reduction), max(f1_macro_reduction))
    
    # Set the y-axis limits to accommodate the max reduction value and a bit more
    plt.ylim([-5, max_reduction + 15])


    # Plot reduction percentages
    line1, = plt.plot(r1, f1_cg_reduction, marker='o', label='F1 (CG) Reduction %', linestyle='-', color='b')
    line2, = plt.plot(r2, f1_macro_reduction, marker='o', label='F1 (Macro) Reduction %', linestyle='-', color='r')

    # Annotate bars with their values
     
    
    for i, (bar, red) in enumerate(zip(bars1, f1_cg_reduction)):
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval + 0.01, round(yval, 4), ha='center', va='bottom', fontsize=16)
        
    for i, (bar, red) in enumerate(zip(bars2, f1_macro_reduction)):
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval + 0.01, round(yval, 4), ha='center', va='bottom', fontsize=16)

    # Annotate lines with reduction percentages
    
        # Annotate lines with absolute change percentages
    for i, change in enumerate(f1_cg_reduction):
        vertical_offset = 2
        if change < 1:  # Adjust this threshold if needed
            vertical_offset = bars1[i].get_height() + 3  # Adjust this offset to place it a bit above the bar
        plt.text(i, change + vertical_offset, f"{round(change, 1)}%", ha='center', color=line1.get_color(), fontsize=18)

    for i, change in enumerate(f1_macro_reduction):
        vertical_offset = 2
        if change < 1:
            vertical_offset = bars2[i].get_height() + 3
        plt.text(i + barWidth, change + vertical_offset, f"{round(change, 1)}%", ha='center', color=line2.get_color(), fontsize=18)


    # Legend font size
    plt.legend(loc='upper left',fontsize=20)
    plt.grid(axis='y', linestyle='--', linewidth=0.5, alpha=0.7)

    plt.tight_layout()
    plt.show()
```
